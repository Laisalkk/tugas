```python
import numpy as np
data = pd.read_csv('/content/penduduk_gender_head.csv')
pivot_table = pd.pivot_table(data, values=["LAKI-LAKI WNI", "PEREMPUAN WNI"], index=["NAMA KECAMATAN", "NAMA KABUPATEN/KOTA",], columns= "NAMA PROVINSI", aggfunc=np.sum)
pivot_table
```

```python
# Membaca data
df = pd.read_csv('/content/penduduk_gender_head.csv')

# 1. Mengelompokkan data berdasarkan 'NAMA PROVINSI'
grouped_by_provinsi = df.groupby('NAMA PROVINSI')
# Menghitung rata-rata, minimum, dan maksimum untuk kolom populasi berdasarkan NAMA PROVINSI
statistik_data = grouped_by_provinsi.agg({
'LAKI-LAKI WNI': ['mean', 'min', 'max'],
'PEREMPUAN WNI': ['mean', 'min', 'max']
})
print("\nStatistik Rata-Rata, Minimum, dan Maksimum Populasi Laki-Laki dan Perempuan WNI per Provinsi:")
print(statistik_data)
```

```python
# Group-wise Aggregation berdasarkan NAMA PROVINSI
grouped_aggregation = df.groupby('NAMA PROVINSI').agg({
'LAKI-LAKI WNI': ['sum', 'min', 'max', 'mean', 'count'],
'PEREMPUAN WNI': ['sum', 'min', 'max', 'mean', 'count'],
'LAKI-LAKI WNA': ['sum', 'min', 'max', 'mean', 'count'],
'PEREMPUAN WNA': ['sum', 'min', 'max', 'mean', 'count']
})

# Menampilkan hasil
print(grouped_aggregation)
```

```python
# Menghitung total populasi Laki-Laki WNI dan Perempuan WNI di setiap provinsi
agg_data = grouped_by_provinsi.agg({
'LAKI-LAKI WNI': 'sum',
'PEREMPUAN WNI': 'sum'
}).reset_index()
print("Agregasi Total Populasi Laki-Laki dan Perempuan WNI per Provinsi:")
print(agg_data)
```

```python
# Menghitung proporsi Laki-Laki WNI di setiap kelurahan berdasarkan total Laki-Laki WNI di provinsi tersebut
df['Proporsi_LAKI_LAKI_WNI'] = grouped_by_provinsi['LAKI-LAKI WNI'].transform(lambda x: x / x.sum())
df['Proporsi_PEREMPUAN_WNI'] = grouped_by_provinsi['PEREMPUAN WNI'].transform(lambda x: x / x.sum())

# Menampilkan data dengan kolom proporsi baru
print("\nData dengan Kolom Proporsi Laki-Laki dan Perempuan WNI per Kelurahan dalam Setiap Provinsi:")
print(df[['NAMA PROVINSI', 'NAMA KELURAHAN', 'LAKI-LAKI WNI', 'Proporsi_LAKI_LAKI_WNI', 'PEREMPUAN WNI', 'Proporsi_PEREMPUAN_WNI']])
```
